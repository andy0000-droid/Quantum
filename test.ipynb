```python
from qiskit import *
from qiskit.providers.aer import AerSimulator
from qiskit.providers.fake_provider import FakeMontreal
import numpy as np
```

```python
backend = FakeMontreal()
```

```python
q_circuit = 3
c_circuit = q_circuit
```

```python
qreg = QuantumRegister(q_circuit,"q")
creg = ClassicalRegister(c_circuit,"c")
circuit_ccx = QuantumCircuit(qreg,creg)
```

```python
circuit_ccx.reset(qreg)
circuit_ccx.x(qreg[0])
circuit_ccx.x(qreg[1])
circuit_ccx.ccx(qreg[0],qreg[1],qreg[2])
```

```python
circuit_ccx.draw(output="mpl")
```

```python
circuit_ccx.decompose().draw("mpl")
```

```python
from qiskit.providers.aer import AerSimulator
sim_Montreal = AerSimulator.from_backend(backend) #Simulator with backend
circuit_basis = transpile(circuit_ccx,sim_Montreal) 
circuit_basis.draw(output="mpl")
```

```python
print("Original depth:",circuit_ccx.depth(),"Decomposed depth:",circuit_ccx.decompose().depth(),"Transpile circuit",circuit_basis.depth())
```
